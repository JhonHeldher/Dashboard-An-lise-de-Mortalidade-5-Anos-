```python
# Usar melt para transformar colunas de meses em uma coluna
df_melt = dfbar3.melt(id_vars="ano", var_name="mes", value_name="valor")

# Criar um dicionário para mapear os nomes dos meses para os números
meses = {"Janeiro": "01", "Fevereiro": "02", "Março": "03","Abril": "04","Maio": "05","Junho": "06","Julho": "07","Agosto": "08","Setembro": "09","Outubro": "10","Novembro": "11","Dezembro": "12",
         }

# Substituir os valores na coluna de mês pelo dicionário
df_melt["mes"] = df_melt["mes"].replace(meses)

# Usar to_datetime para criar uma coluna de data a partir de ano e mês
df_melt["data"] = df_melt["ano"].astype(str) + "-" + df_melt["mes"]

# Excluir as colunas de ano e mês
df_melt = df_melt.drop(columns=["ano", "mes"])

df_melt = df_melt.sort_values("data")

df_melt = df_melt.reset_index(drop=True)
```

```python
# Usar melt para transformar colunas de meses em uma coluna
df_melt2 = dfbar4.melt(id_vars="ano", var_name="mes", value_name="valor")

# Substituir os valores na coluna de mês pelo dicionário
df_melt2["mes"] = df_melt2["mes"].replace(meses)

# Usar to_datetime para criar uma coluna de data a partir de ano e mês
df_melt2["data"] = df_melt2["ano"].astype(str) + "-" + df_melt2["mes"]

# Excluir as colunas de ano e mês
df_melt2 = df_melt2.drop(columns=["ano", "mes"])

df_melt2 = df_melt2.sort_values("data")

df_melt2 = df_melt2.reset_index(drop=True)
```

```python
df_melt3 = df_melt.groupby('data')['valor'].sum() - df_melt2.groupby('data')['valor'].sum()
df_melt3 = df_melt3.reset_index(name='valor')
# df_melt3 = df_melt3.query("'2020-01' <= data <= '2022-12'")
```
